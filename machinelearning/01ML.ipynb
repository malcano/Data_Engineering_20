```python
#학습용 데이터와 검증용 데이터를 따로 구분해놓아야 함 
from sklearn.model_selection import train_test_split

#psedo seed number 랜덤 시드 넘버를 결정
```

```python
# 변수명도 관습적으로 학습, 검증을 아래처럼 구분함 
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, \
                                                    random_state=42)
```

```python
train_test_split(iris.data, iris.target, test_size=0.3, \
                                                    random_state=42)
#꽃의 데이터를 7대 3으로 나눔
```

```python
len(X_test)
```

```python
#데이터 분할을 했고 결정 트리 모델 함 해보자
from sklearn.tree import DecisionTreeClassifier
# 1) 분석을 수행하기 위한 추정치(estimator)생성
cls = DecisionTreeClassifier()

# 2) 해당 객체에 대해 fit()메소드를 수행: 학습수행
cls.fit(X_train, y_train)

# 3) 검증하고 예측 결과 수행 
y_pred = cls.predict(X_test)
# 얘가 값을 내놓을텐데 일단 예를 받아놓아라
y_pred
```

```python
# 예측한 것 가지고 몇 개가 맞았나를 보면 됨 
# 모델 평가 
# 생성된 객처 또는 수행된 분석 결과에 대해 적합한 성능 평가 지표 도출
# 검증할 때는 accuracy_score 씀
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
# 1.0 == 100점, 즉 잘 만들어진 모델이다
```

```python
#10대 20대 30대 40대 구분을 해봅시다 
def get_age_cat(x):
    if x<=5:
        cat = 'Baby'
    elif x<=12:
        cat = 'Child'
    elif x<=18:
        cat = 'Teenager'
    elif x<=25:
        cat = 'Student'
    elif x<=35:
        cat = 'Young Adult'
    elif x<=60:
        cat = 'Adult'
    else:
        cat = 'Elderly'
    return cat
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import numpy as np

items =  ['TV', '냉장고','전자렌지', '컴퓨터', '선풍기', '선풍기', '믹서','믹서']
encoder  = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)

# 원핫 인코딩 하기 전 2차원 배열로 구조 변경
labels = labels.reshape((-1,1)) #resahpe: 구조의 모양을 tuple로 넣음 // 몇 행일지는 모르겠지만 한 행에는 무조건 데이터 하나만 들어가야 함 

oh_encoder = OneHotEncoder()
oh_encoder.fit(labels)
oh_labels = oh_encoder.transform(labels)
oh_labels.toarray()
```
