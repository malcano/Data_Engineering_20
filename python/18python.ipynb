xlim(최소값, 최대값)과 ylim(최소값, 최대값)을 통해 축의 최소값과 최대값을 지정할 수 있다

xticks()이나 yticks()를 사용해서 맷플롯리브 수동으로 설정할 수 있다

```python
import numpy as np

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)

plt.title("X축과 Y축의 tick label 설정" )
plt.plot(X,C)
plt.xticks([-np.pi, -np.pi / 2, 0 , np.pi / 2, np.pi])
plt.yticks([-1,0,+1])
plt.show()
```

틱라벨 문자열에는 $$사이에 수학 문자식을 넣을 수도 있다

https://matplotlib.org/3.1.1/tutorials/text/mathtext.html

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)

plt.title("X축과 Y축의 tick label 설정")
plt.plot(X, C)
plt.yticks([-1, 0, +1],["Low","Zero","High"])
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$\pi$', r'$\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
plt.show()
```

### Axes

Figure 객체 안에 여러 개의 플롯을 그리드 형태로 보여야 하는 경우도 있음

subplot()을 사용하여 명시적으로 Axes객체를 얻어야 함

plot()을 바로 사용해도 자동으로 Figure Axes객체를 생성해 준다. 

subplot()은 그리드 형태의 Axes객체를 생성한다. Figure가 행렬이고 Axes가 행렬의 원소라고 생각하면 된다 .
